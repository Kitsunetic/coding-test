1차원 DP. 2차원의 경우 대놓고 DP라고 보이니깐 괜찮은데, 1차원이면 DP인지 아닌지 햇갈릴 때가 있어서 가끔 당황시키는 문제

DP는 초기 값과, 확정적인(=가장 가까운) 값을 먼저 정해놓고, 그 값들을 활용해서 계속해서 이어지는 다음 값들을 채워 나가는 방법.

"최솟값이 8보다 크면 -1을 return 합니다." 가 중요한 힌트

```python
N, number = 2, 10
```

```python
def arithmetic(A, B):
    S = set()
    for a in A:
        for b in B:
            for v in (a + b, a - b, a * b, a // b, b - a, b // a):
                if 0 < v <= 32000:
                    S.add(v)

    return S
```

```python
def arithmetic2(A, B):
    S = set()
    for a in A:
        for b in B:
            for v in (a + b, a - b, a * b, a // b, b - a, b // a):
                if 0 < v:  # <= 32000:
                    S.add(v)

    return S
```

```python
dp = [set() for _ in range(10)]
```

단순히 N을 쌓아서 만들 수 있는 값들 추가

```python
dp = [set() for _ in range(9)]

for i, m in enumerate([1, 11, 111, 1111, 11111], 1):
    v = N * m
    dp[i].add(v)

for d in range(2, 9):
    for i in range(1, d // 2 + 1):
        S = arithmetic(dp[i], dp[d - i])

        dp[d].update(S)
```

```python
dp1 = dp
```

```python
dp = [set() for _ in range(9)]
dp[1].add(N)

for i in range(2, 9):
    v = int(f"{N}" * i)
    dp[i].add(v)

    for j in range(1, i // 2 + 1):
        S = arithmetic2(dp[j], dp[i - j])

        dp[i].update(S)
```

```python
dp2 = dp
```
